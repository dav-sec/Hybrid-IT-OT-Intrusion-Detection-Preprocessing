```python
# ============================================
# CELDA 1: IMPORTS
# ============================================

import pandas as pd
import numpy as np

from pathlib import Path
```

```python
# ============================================
# CELDA 2: DIRECTORIOS
# ============================================

BASE_DIR = Path.cwd()

PHASE1_DIR = BASE_DIR / "output_phase1"

OUTPUT_DIR = BASE_DIR / "output_phase2"

OUTPUT_DIR.mkdir(exist_ok=True)

FLOW_CSV = PHASE1_DIR / "raw_flow_features.csv"

LOG_CSV = PHASE1_DIR / "raw_log_features.csv"

FLOW_OUTPUT = OUTPUT_DIR / "clean_flow_features.csv"

LOG_OUTPUT = OUTPUT_DIR / "clean_log_features.csv"

FUSED_OUTPUT = OUTPUT_DIR / "fused_it_ot.csv"
```

```python
# ============================================
# CELDA 3: VALIDAR ARCHIVOS
# ============================================

files = [

    FLOW_CSV,
    LOG_CSV

]

for file in files:

    print()

    print(f"VALIDANDO: {file.name}")

    if file.exists():

        print("EXISTE")

    else:

        raise FileNotFoundError(

            f"{file.name} NO EXISTE"

        )
```

```python
# ============================================
# CELDA 4: LIMPIEZA FLOW DATASET
# ============================================

chunk_counter = 0

for chunk in pd.read_csv(

    FLOW_CSV,

    chunksize=100000

):

    chunk_counter += 1

    print()

    print(f"PROCESANDO FLOW CHUNK {chunk_counter}")

    original_rows = len(chunk)

    # ============================================
    # ELIMINAR DUPLICADOS
    # ============================================

    duplicates = chunk.duplicated().sum()

    if duplicates > 0:

        print(f"DUPLICADOS DETECTADOS: {duplicates}")

        chunk = chunk.drop_duplicates()

    else:

        print("SIN DUPLICADOS")

    # ============================================
    # ELIMININAR NULOS CRÍTICOS
    # ============================================

    critical_columns = [

        "timestamp",
        "src_ip",
        "dst_ip",
        "protocol",
        "traffic_label"

    ]

    nulls = chunk[critical_columns].isnull().sum().sum()

    if nulls > 0:

        print(f"NULOS CRÍTICOS: {nulls}")

        chunk = chunk.dropna(

            subset=critical_columns

        )

    else:

        print("SIN NULOS CRÍTICOS")

    # ============================================
    # TIMESTAMP
    # ============================================

    chunk["timestamp"] = pd.to_datetime(

        chunk["timestamp"],

        unit="ms",

        errors="coerce"

    )

    invalid_timestamps = (

        chunk["timestamp"]
        .isnull()
        .sum()

    )

    if invalid_timestamps > 0:

        print(

            f"TIMESTAMPS INVÁLIDOS: {invalid_timestamps}"

        )

        chunk = chunk.dropna(

            subset=["timestamp"]

        )

    # ============================================
    # EXPORTAR
    # ============================================

    if chunk_counter == 1:

        chunk.to_csv(

            FLOW_OUTPUT,

            index=False,

            mode="w"

        )

    else:

        chunk.to_csv(

            FLOW_OUTPUT,

            index=False,

            mode="a",

            header=False

        )

    print()

    print(

        f"ROWS ORIGINALES: {original_rows}"

    )

    print(

        f"ROWS LIMPIAS: {len(chunk)}"

    )

    del chunk
```

```python
# ============================================
# CELDA 5: LIMPIEZA LOG DATASET
# ============================================

chunk_counter = 0

for chunk in pd.read_csv(

    LOG_CSV,

    chunksize=100000

):

    chunk_counter += 1

    print()

    print(f"PROCESANDO LOG CHUNK {chunk_counter}")

    original_rows = len(chunk)

    # ============================================
    # DUPLICADOS
    # ============================================

    duplicates = chunk.duplicated().sum()

    if duplicates > 0:

        print(f"DUPLICADOS DETECTADOS: {duplicates}")

        chunk = chunk.drop_duplicates()

    else:

        print("SIN DUPLICADOS")

    # ============================================
    # NULOS CRÍTICOS
    # ============================================

    critical_columns = [

        "raw_log",
        "traffic_label"

    ]

    nulls = chunk[critical_columns].isnull().sum().sum()

    if nulls > 0:

        print(f"NULOS CRÍTICOS: {nulls}")

        chunk = chunk.dropna(

            subset=critical_columns

        )

    else:

        print("SIN NULOS CRÍTICOS")

    # ============================================
    # EXPORTACIÓN
    # ============================================

    if chunk_counter == 1:

        chunk.to_csv(

            LOG_OUTPUT,

            index=False,

            mode="w"

        )

    else:

        chunk.to_csv(

            LOG_OUTPUT,

            index=False,

            mode="a",

            header=False

        )

    print()

    print(

        f"ROWS ORIGINALES: {original_rows}"

    )

    print(

        f"ROWS LIMPIAS: {len(chunk)}"

    )

    del chunk
```

```python
# ============================================
# CELDA 6: CARGAR DATASETS LIMPIOS
# ============================================

flow_df = pd.read_csv(

    FLOW_OUTPUT,

    nrows=500000

)

log_df = pd.read_csv(

    LOG_OUTPUT,

    nrows=500000

)

print()

print("FLOW DATASET")

print(flow_df.shape)

print()

print("LOG DATASET")

print(log_df.shape)
```

```python
# ============================================
# CELDA 7: SINCRONIZACIÓN IT/OT
# ============================================

# ============================================
# GENERAR ÍNDICES TEMPORALES
# ============================================

flow_df = flow_df.reset_index(drop=True)

log_df = log_df.reset_index(drop=True)

# ============================================
# TAMAÑO MÍNIMO
# ============================================

min_rows = min(

    len(flow_df),

    len(log_df)

)

flow_df = flow_df.iloc[:min_rows]

log_df = log_df.iloc[:min_rows]

print()

print(f"ROWS SINCRONIZADAS: {min_rows}")
```

```python
# ==========================================
# CELDA 8 — FUSIÓN IT/OT LIMPIA
# ==========================================

flow_df = pd.read_csv(FLOW_OUTPUT)
log_df = pd.read_csv(LOG_OUTPUT)

# sincronizar cantidad de filas
min_rows = min(len(flow_df), len(log_df))

flow_df = flow_df.iloc[:min_rows].reset_index(drop=True)
log_df = log_df.iloc[:min_rows].reset_index(drop=True)

print(f"ROWS SINCRONIZADAS: {min_rows}")

# eliminar columnas duplicadas del dataset LOG
log_df = log_df.drop(columns=[
    "attack_type",
    "traffic_label"
], errors="ignore")

# fusionar IT + OT
fused_df = pd.concat([flow_df, log_df], axis=1)

print("\nDATASET FUSIONADO")
print(fused_df.shape)

print("\nCOLUMNAS FINALES")
print(fused_df.columns)

print("\nMUESTRA DATASET")
fused_df.head()
```

```python
# ============================================
# CELDA 9: CORREGIR traffic_label + EXPORTAR
# ============================================

import numpy as np

# ── Corrección binaria de traffic_label ──────────────────────
BENIGN_LABEL = "normal_pcaps"

fused_df["attack_type"] = fused_df["attack_type"].astype(str)

fused_df["traffic_label"] = np.where(
    fused_df["attack_type"].str.strip() == BENIGN_LABEL,
    0,
    1
)

# ── Validación ───────────────────────────────────────────────
print("=" * 55)
print("DISTRIBUCIÓN traffic_label RECALCULADO")
print("=" * 55)

dist = fused_df["traffic_label"].value_counts()
total = len(fused_df)

for label, count in dist.items():
    tipo = "BENIGNO  (normal_pcaps)" if label == 0 else "ATAQUE   (todos los subtipos)"
    pct = count / total * 100
    print(f"  {label} → {tipo:<30}  {count:>8,}  ({pct:.2f}%)")

print("-" * 55)
print(f"  TOTAL filas: {total:>8,}")

leakage = fused_df[
    (fused_df["attack_type"].str.strip() == BENIGN_LABEL) &
    (fused_df["traffic_label"] == 1)
]
assert len(leakage) == 0, f"ERROR: {len(leakage)} filas benignas etiquetadas como ataque!"
print("\n  Verificación cruzada: SIN fuga de etiquetas ✓")
print("=" * 55)

# ── Exportar ─────────────────────────────────────────────────
fused_df.to_csv(
    FUSED_OUTPUT,
    index=False
)

print()
print("DATASET FINAL EXPORTADO")
print(FUSED_OUTPUT)
```
