```python
# ============================================
# CELDA 1: IMPORTS
# ============================================

import pandas as pd
import numpy as np

from pathlib import Path

import warnings

warnings.filterwarnings("ignore")
```

```python
# ============================================
# CELDA 2: CONFIGURACIÓN DE DIRECTORIOS
# ============================================

BASE_DIR = Path.cwd()

PHASE1_OUTPUT = BASE_DIR / "output_phase1"

FLOW_CSV = PHASE1_OUTPUT / "raw_flow_features.csv"

LOG_CSV = PHASE1_OUTPUT / "raw_log_features.csv"

WINDOW_CSV = PHASE1_OUTPUT / "sliding_window_features.csv"

print("ARCHIVOS DETECTADOS:")

print()

print(FLOW_CSV)

print(LOG_CSV)

print(WINDOW_CSV)
```

```python
# ============================================
# CELDA 3: VALIDACIÓN DE EXISTENCIA
# ============================================

files_to_validate = [

    FLOW_CSV,
    LOG_CSV,
    WINDOW_CSV

]

for file in files_to_validate:

    print()

    print(f"VALIDANDO: {file.name}")

    if file.exists():

        print("EXISTE")

    else:

        raise Exception(

            f"NO EXISTE: {file}"

        )
```

```python
# ============================================
# CELDA 4: TAMAÑO DE ARCHIVOS
# ============================================

for file in files_to_validate:

    size_gb = (

        file.stat().st_size / (1024**3)

    )

    print()

    print(file.name)

    print(f"{size_gb:.2f} GB")
```

```python
# ============================================
# CELDA 5: VALIDACIÓN DE HEADER
# ============================================

print("===================================")
print("FLOW HEADER")
print("===================================")

flow_header = pd.read_csv(

    FLOW_CSV,
    nrows=5

)

print(flow_header.columns)

print()

print("===================================")
print("LOG HEADER")
print("===================================")

log_header = pd.read_csv(

    LOG_CSV,
    nrows=5

)

print(log_header.columns)

print()

print("===================================")
print("WINDOW HEADER")
print("===================================")

window_header = pd.read_csv(

    WINDOW_CSV,
    nrows=5

)

print(window_header.columns)
```

```python
# ============================================
# CELDA 6: VALIDACIÓN DE MUESTRAS
# ============================================

print("===================================")
print("FLOW SAMPLE")
print("===================================")

flow_sample = pd.read_csv(

    FLOW_CSV,
    nrows=5

)

print(flow_sample.head())

print()

print("===================================")
print("LOG SAMPLE")
print("===================================")

log_sample = pd.read_csv(

    LOG_CSV,
    nrows=5

)

print(log_sample.head())

print()

print("===================================")
print("WINDOW SAMPLE")
print("===================================")

window_sample = pd.read_csv(

    WINDOW_CSV,
    nrows=5

)

print(window_sample.head())
```

```python
# ============================================
# CELDA 7: VALIDACIÓN DE COLUMNAS CRÍTICAS
# ============================================

required_flow_columns = [

    "timestamp",
    "src_ip",
    "dst_ip",
    "protocol",
    "attack_type",
    "traffic_label"

]

required_log_columns = [

    "raw_log",
    "device_type",
    "attack_type",
    "traffic_label"

]

required_window_columns = [

    "timestamp",
    "packet_count",
    "byte_count",
    "inter_arrival_time",
    "jitter"

]

print("===================================")
print("VALIDANDO FLOW COLUMNS")
print("===================================")

missing_flow = [

    col for col in required_flow_columns

    if col not in flow_sample.columns

]

if len(missing_flow) == 0:

    print("FLOW OK")

else:

    print("FALTAN:")

    print(missing_flow)

print()

print("===================================")
print("VALIDANDO LOG COLUMNS")
print("===================================")

missing_log = [

    col for col in required_log_columns

    if col not in log_sample.columns

]

if len(missing_log) == 0:

    print("LOG OK")

else:

    print("FALTAN:")

    print(missing_log)

print()

print("===================================")
print("VALIDANDO WINDOW COLUMNS")
print("===================================")

missing_window = [

    col for col in required_window_columns

    if col not in window_sample.columns

]

if len(missing_window) == 0:

    print("WINDOW OK")

else:

    print("FALTAN:")

    print(missing_window)
```

```python
# ============================================
# CELDA 8: VALIDACIÓN DE LABELS
# ============================================

print("===================================")
print("FLOW LABELS")
print("===================================")

flow_labels = pd.read_csv(

    FLOW_CSV,
    usecols=["traffic_label"],
    nrows=100000

)

print(

    flow_labels["traffic_label"]
    .value_counts()

)

print()

print("===================================")
print("FLOW ATTACK TYPES")
print("===================================")

flow_attacks = pd.read_csv(

    FLOW_CSV,
    usecols=["attack_type"],
    nrows=100000

)

print(

    flow_attacks["attack_type"]
    .value_counts()

)

print()

print("===================================")
print("LOG ATTACK TYPES")
print("===================================")

log_attacks = pd.read_csv(

    LOG_CSV,
    usecols=["attack_type"],
    nrows=100000

)

print(

    log_attacks["attack_type"]
    .value_counts()

)
```

```python
# ============================================
# CELDA 10: VALIDACIÓN DE DUPLICADOS
# ============================================

print("===================================")
print("FLOW DUPLICADOS")
print("===================================")

flow_duplicates = pd.read_csv(

    FLOW_CSV,
    nrows=100000

)

flow_dup_count = flow_duplicates.duplicated().sum()

print(flow_dup_count)

print()

print("===================================")
print("LOG DUPLICADOS")
print("===================================")

log_duplicates = pd.read_csv(

    LOG_CSV,
    nrows=100000

)

log_dup_count = log_duplicates.duplicated().sum()

print(log_dup_count)

print()

print("===================================")
print("WINDOW DUPLICADOS")
print("===================================")

window_duplicates = pd.read_csv(

    WINDOW_CSV,
    nrows=100000

)

window_dup_count = window_duplicates.duplicated().sum()

print(window_dup_count)
```

```python
# ============================================
# CELDA 11: VALIDACIÓN DE TIMESTAMPS
# ============================================

print("===================================")
print("FLOW TIMESTAMPS")
print("===================================")

flow_time = pd.read_csv(

    FLOW_CSV,
    usecols=["timestamp"],
    nrows=100000

)

flow_time["timestamp"] = pd.to_datetime(

    flow_time["timestamp"],
    errors="coerce",
    unit="ms"

)

print(flow_time.head())

print()

print("TIMESTAMPS INVÁLIDOS:")

print(

    flow_time["timestamp"]
    .isnull()
    .sum()

)

print()

print("===================================")
print("WINDOW TIMESTAMPS")
print("===================================")

window_time = pd.read_csv(

    WINDOW_CSV,
    usecols=["timestamp"],
    nrows=100000

)

window_time["timestamp"] = pd.to_datetime(

    window_time["timestamp"],
    errors="coerce"

)

print(window_time.head())

print()

print("TIMESTAMPS INVÁLIDOS:")

print(

    window_time["timestamp"]
    .isnull()
    .sum()

)
```

```python
# ============================================
# CELDA EXTRA: VALIDACIÓN GLOBAL DE CATEGORÍAS
# ============================================

import pandas as pd

attack_counts = {}

chunk_counter = 0

for chunk in pd.read_csv(

    FLOW_CSV,

    usecols=[

        "attack_type",
        "traffic_label"

    ],

    chunksize=100000

):

    chunk_counter += 1

    print(f"PROCESANDO CHUNK {chunk_counter}")

    counts = (

        chunk["attack_type"]
        .value_counts()

    )

    for attack, count in counts.items():

        attack_counts[attack] = (

            attack_counts.get(attack, 0)

            + count

        )

print()

print("===================================")
print("CATEGORÍAS GLOBALES")
print("===================================")

for attack, count in attack_counts.items():

    print(f"{attack}: {count}")
```

```python
# ============================================
# CELDA EXTRA: VALIDACIÓN GLOBAL LOGS
# ============================================

log_attack_counts = {}

chunk_counter = 0

for chunk in pd.read_csv(

    LOG_CSV,

    usecols=[

        "attack_type"

    ],

    chunksize=100000

):

    chunk_counter += 1

    print(f"PROCESANDO CHUNK {chunk_counter}")

    counts = (

        chunk["attack_type"]
        .value_counts()

    )

    for attack, count in counts.items():

        log_attack_counts[attack] = (

            log_attack_counts.get(attack, 0)

            + count

        )

print()

print("===================================")
print("LOG CATEGORÍAS GLOBALES")
print("===================================")

for attack, count in log_attack_counts.items():

    print(f"{attack}: {count}")
```
