```python
import os
import joblib
import pandas as pd
from sklearn.preprocessing import LabelEncoder

FUSED_PATH = "fused_it_ot.csv"
CARPETA_MODELO = "modelo_guardado"

os.makedirs(CARPETA_MODELO, exist_ok=True)

categorical_columns = [
    "application_name",
    "application_category",
    "device_type"
]

print("COLUMNAS CATEGORICAS PARA IDS:")
print(categorical_columns)

encoders = {}

for column in categorical_columns:
    print(f"\nENTRENANDO ENCODER: {column}")

    unique_values = set()

    reader = pd.read_csv(
        FUSED_PATH,
        usecols=[column],
        chunksize=100000,
        low_memory=False
    )

    for chunk in reader:
        chunk[column] = (
            chunk[column]
            .astype(str)
            .str.strip()
            .str.lower()
            .fillna("unknown")
        )
        unique_values.update(chunk[column].unique())

    encoder = LabelEncoder()
    encoder.fit(sorted(list(unique_values)))

    encoders[column] = encoder

    print(f"CLASES DETECTADAS: {len(encoder.classes_)}")

# Guardar encoders
ruta_encoders = os.path.join(CARPETA_MODELO, "encoders.pkl")
joblib.dump(encoders, ruta_encoders)

print(f"\nENCODERS GUARDADOS EN: {ruta_encoders}")
```
